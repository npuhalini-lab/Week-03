```python
#if, elif and else

marks=int(input("Enter the marks"))
if marks>=75:
    print("Grade A")
elif marks>=65:
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Grade F")
```

```python
#+ve, -ve numbers - if and else

number=int(input("Enter the number"))
if number>0:
    print("Positive Number")
elif number<0:
    print("Negative Number")
else:
    print("Zero")
```

```python
#+ve, -ve numbers - if, elif and else (nested)

number=int(input("Enter the number"))
if number>0:
    print("Positive Number")
else:
    if number<0:
        print("Negative Number")
    else:
        print("Zero")
```

```python
## Question 03
name=input("Enter the name")
marks=int(input("Enter the marks"))

if 100>=marks>=85:
    print("Excellent",name, "You have A pass")
elif marks>=75:
    print("Great",name, "You have B pass")
elif marks>=65:
    print("Good",name, "You have C pass")
elif marks>=50:
    print("Fair",name, "You have D pass")
else:
    print("Sorry",name, "You have failed. Try again")
```
